##### Runge–Kutta–2 (RK2) to solve System of ODEs

$$
\boldsymbol{r}_{n+1}
= \boldsymbol{r}_n + \frac{h}{2}\,(\boldsymbol{K_1} + \boldsymbol{K}_2)
$$

where
$$
\begin{aligned}
\boldsymbol{K}_1 &= \boldsymbol{F}(t_n,\boldsymbol{r}_n) \\
\boldsymbol{K}_2 &= \boldsymbol{F}\!\left(t_n+h,\,
\boldsymbol{r}_n+h\boldsymbol{K}_1\right)
\end{aligned}
$$

with local error $O(h^3)$ and global error $O(h^2)$.


```python
import numpy as np
def F(t, Y):
    x, y = Y
    dxdt = y
    dydt = - x
    return np.array([dxdt, dydt])

# RK 2
def RK2(F, t0, Y0, tn, n=10):
    ts = np.array([t0])
    Ys = np.array([Y0])
    dt = (tn-t0)/n

    for _ in range(n):
        ti, Yi = ts[-1], Ys[-1,:]
        K1 = F(ti, Yi)
        K2 = F(ti + dt, Yi + dt * K1)
        Y = Yi + dt/2 * (K1 + K2)
        Ys = np.append(Ys, [Y], axis=0)
        ts = np.append(ts, ti + dt)
        
    return ts, Ys

ts, Ys = RK2(F, 0, [1, 0], 2, 20)
```

##### Runge-Kutta-4 (RK4)
$$
\bold{r}_{n+1} = \bold{r}_n + \frac{h}{6} \space (K_1 + 2K_2 + 2K_3 + K_4)
$$
where \
$
\bold{K_1} = \bold{F}(t_n, \bold{r}_n) \\
\bold{K_2} = \bold{F}(t_n + \frac{h}{2}, \bold{r}_n + \frac{h}{2}\bold{K_1}) \\
\bold{K_3} = \bold{F}(t_n + \frac{h}{2}, \bold{r}_n + \frac{h}{2}\bold{K_2}) \\
\bold{K_4} = \bold{F}(x_n + h, \bold{r}_n + h \bold{K_3}) \\
$\
with local error $O(h^5)$ but Global Error $O(h^4)$

```python
import numpy as np
def F(t, Y):
    x, y = Y
    dxdt = y
    dydt = - x
    return np.array([dxdt, dydt])

# RK4
def RK4(F, t0, Y0, tn, n=10):
    ts = np.array([t0])
    Ys = np.array([Y0])
    dt = (tn-t0)/n

    for _ in range(n):
        ti, Yi = ts[-1], Ys[-1,:]
        K1 = F(ti, Yi)
        K2 = F(ti + dt/2, Yi + dt/2 * K1)
        K3 = F(ti + dt/2, Yi + dt/2 * K2)
        K4 = F(ti + dt, Yi + dt * K3)
        Y = Yi + dt/6 * (K1 + 2*K2 + 2*K3 +K4)
        Ys = np.append(Ys, [Y], axis=0)
        ts = np.append(ts, ti + dt)
    return ts, Ys

ts, Ys = RK4(F, 0, [1, 0], 2, 20)
```

```python
import matplotlib.pyplot as plt
# solve Y
def F(t, Y):
    x, y, z = Y
    sigma, rho, beta = 10, 28, 8/3
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta*z
    return np.array([dxdt, dydt, dzdt])

ts, Ys = RK2(F, 0, [1,1,1], 30, 5000)

# plot x(t)
plt.figure(figsize=(4, 3))
plt.title('x(t)')
plt.plot(ts, Ys[:,0], '-', label='rk4')
#
plt.grid()
plt.legend()
plt.show()

#plot y(t)
plt.figure(figsize=(4, 3))
plt.title('y(t)')
plt.plot(ts, Ys[:,1], '-', label='rk4')
#
plt.grid()
plt.legend()
plt.show()

#plot z(t)
plt.figure(figsize=(4, 3))
plt.title('z(t)')
plt.plot(ts, Ys[:,2], '-', label='rk4')
#
plt.grid()
plt.legend()
plt.show()
```
