```python
import matplotlib.pyplot as plt
import numpy as np

# datas
x_num = np.array([0, 0.25, 0.5])
y_euler = np.array([1.0, 1.0, 0.875])
y_mid = np.array([1.0, 0.9375, 0.7915])
y_rk2 = np.array([1.0, 0.9375, 0.7969])
y_rk4 = np.array([1.0, 0.941154, 0.800065])

# exact
x_exact = np.linspace(0, 0.5, 100)
y_exact = 1 / (x_exact**2 + 1)

# Plot exact
plt.plot(x_exact, y_exact, 'k-', label='Exact: $1/(x^2+1)$', linewidth=2)

# Plot approx
plt.plot(x_num, y_euler, 'ro--', label='Euler')
plt.plot(x_num, y_mid, 'go--', label='Midpoint')
plt.plot(x_num, y_rk2, 'bo--', label='RK2 (Heun)')
plt.plot(x_num, y_rk4, 'mo--', label='RK4')

# Label
plt.title('Compare each approx outcome to exact')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.xlim(0, 0.55)
plt.ylim(0.75, 1.05)

plt.show()
```
