## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6, 4))

x = np.arange(0, 10, 0.1) # also can use `np.linspace(0,10,100)`
y = np.sin(x)

# plot the data 
plt.plot(x, y, '^', color='r', label='sin(x)') # color, label, linestyle

# limit the x-axis
plt.xlim(0, 2*np.pi)

plt.title(r'y=$sin(x)$')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x1 = [1, 1]
y1 = [0, 4]
plt.plot(x1, y1, color='k')
plt.plot([1, 1.9], [2, 4], color='k')
plt.plot([1, 1.9], [2, 0], color='k')
plt.plot([2,2,3,4,4], [0,4,0,4,0], color='k')
plt.plot([5,5], [0,4], color='b')
plt.plot([6,7,7,6,6,7], [0,0,2,2,4,4], color='b')
plt.plot([9, 8, 8, 9, 9, 8.5], [4,4,0,0,2,2], color='red')
plt.plot([10, 10.5, 11], [0, 4, 0], color='red')
plt.plot([10.25, 10.75], [2,2], color='red')
plt.plot([12, 12.5, 12.5], [4, 2, 0], color='red')
plt.plot([12.5, 13], [2, 4], color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d


# create a new figure for plotting
fig = plt.figure(figsize=(4, 3))

# create a new subplot on our figure
# and set projection as 3d
ax1 = fig.add_subplot(1,1,1, projection='3d')

# defining x, y, z co-ordinates
x = np.random.randint(0, 10, size = 20)
y = np.random.randint(0, 10, size = 20)
z = np.random.randint(0, 10, size = 20)

# plotting the points on subplot
ax1.scatter(x,y,z,color='b',marker='o')

# setting labels for the axes
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
ax1.set_zlabel('z-axis')

plt.title("3d plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6,1) #6 is exclusive, len = final-initial
y = np.array([3,4,5,8,6])

plt.bar(x, y, width=1,ccolor='lightgreen', label='histogram')

plt.title("Histogram")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()

```
