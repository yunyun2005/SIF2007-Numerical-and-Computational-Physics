```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\user\Downloads\lsdata1.xlsx")

x_samp = df['Temperature']
y_samp = df['Conductance']

param = np.polyfit(x_samp, y_samp, 1)

x_dom = np.arange(100, 1000, 0.5)
y_ran = np.polyval(param, x_dom)

plt.plot(x_dom, y_ran)
plt.plot(x_samp, y_samp, 'o')
plt.xticks(np.arange(100, 1001, 100))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
