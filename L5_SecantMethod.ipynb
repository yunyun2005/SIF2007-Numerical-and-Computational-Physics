### Secant Method
- Similiar as like Newton Raphson but no need df/dx
- Replace df/dx with gradient y/x

```python
def secant(f, x0, x1, tolerance = 1e-6):
# f: func | df: 1st-derivative | x: guess

    x_new = x1 - f(x1) * (x1 - x0)/(f(x1)-f(x0))
    
    if abs(x_new - x1) < tolerance:
        return x_new
    else:
         return secant(f, x1, x_new, tolerance)
```
