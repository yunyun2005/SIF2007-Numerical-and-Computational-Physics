### Piecewise Linear Interpolation
$$ 
p(x)=\frac{x_1-x}{x_1-x_0} f(x_0)+\frac{x-x_0}{x_1-x_0}f(x_1)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
def P(f, x0, x1, x):             # interpolation formula
    L0 = (x1 - x) / (x1 - x0)
    L1 = (x - x0) / (x1 - x0)
    return L0 * f(x0) + L1 * f(x1)

def f(x):                       # function f
    return np.exp(x)

x1 = np.linspace(0, 0.5, 100)   # interval-1
x2 = np.linspace(0.5, 1, 100)   # interval-2

y1 = P(f, 0, 0.5, x1)           # interpolation-1
y2 = P(f, 0.5, 1, x2)           # interpolation-2

plt.plot(x1, y1, color='b', label='interpolation')  # plot interpolation
plt.plot(x2, y2, color='b')
plt.plot([0, 0.5, 1], f(np.array([0, 0.5, 1])), 'o', color='k', label='nodes') # plot nodes

x = np.linspace(0, 1, 100)       # plot exp(x)
plt.plot(x, f(x), label='exp(x)', color='orange')

plt.grid()
plt.legend()
plt.title('Piecewise Interpolation')
plt.ylabel('y')
plt.xlabel('x')
```
