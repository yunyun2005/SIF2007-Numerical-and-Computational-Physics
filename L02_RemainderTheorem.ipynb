### Remainder Theorem

**Remainder theorem** states that for a chopped taylor series, there is a remainder such that
$$
f(x)=f(x_o) + \sum_{n=1}^n{\frac{f^{(n)}(x_o)}{n!}(x-x_o)^n} + R_n(x)
$$

$$
R_n(x) = \frac{f^{(n+1)}(\xi)}{(n+1)!}(x-x_o)^{n+1}
$$
equivalently
$$
R_n(x)=O\Big((x-x_o)^{n+1}\Big)
$$
