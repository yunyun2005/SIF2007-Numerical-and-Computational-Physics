### Numerical Integration

##### Principle:
$$
I \approx f_{0.5} \Delta x + f_{1.5} \Delta x + f_{2.5} \Delta x +...

##### Trapezoid Rule
$$
T_n(f)=\frac{h}{2} \space \sum_{i=0}^{n-1} {f(x_{i})+f(x_{i+1})}
$$ \
with error
$$
I - T_n(f) = -\frac{h^2}{12} (b-a) f''(\xi)

```python
def trapezoid(f, a, b, n=10):
    h = (b - a) / n
    total = 0
    xi = a
    for i in range(n):
        total += (f(xi)+f(xi+h))/2
        xi += h
    return h * total

def f(x):
    return 3 * x**2

trapezoid(f, 0, 5, 1000)
```

##### Midpoint Rule
$$
\begin{aligned}
M_n(f)
&= h \space \sum_{i=0}^{n-1} f\!\left( \frac{x_i + x_{i+1}}{2} \right) \\\\
&= h \space \sum_{i=0}^{n-1} f\!\left( x_i + \frac{h}{2} \right)
\end{aligned}
$$


with error
$$
I - M_n(f) = +\frac{h^2}{24} (b-a)f''(\xi)
$$


```python
def midpoint(f, a, b, n=10):
    h = (b-a) / n
    total = 0
    xi = a
    
    for i in range(n):
        total += f(xi + h/2)
        xi += h
    
    return h * total

def f(x):
    return 3 * x**2

midpoint(f, 0, 5, 1000)
```

##### Simpson's Rule
$$
S_n(f) = \frac{h}{3} \space \sum_{i=1}^{n / 2} \Big[f(x_{2i-2}) + 4f(x_{2i-1}) + f(x_{2i}) \Big]\\
$$

```python
import numpy as np

def simpson(f, a, b, n=10):
    if n % 2 != 0:
        raise Exception("n must be even")

    h = (b - a) / n
    x = np.linspace(a, b, n+1)

    total = 0
    for i in range(int(n/2)):
        i += 1
        x1, x2, x3 = x[2*i - 2], x[2*i - 1], x[2*i]
        total += f(x1) + 4*f(x2) + f(x3)

    return h/3 * total
simpson(f, 0, 5, 100)
```
