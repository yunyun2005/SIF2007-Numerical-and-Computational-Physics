```python
def findmin(f, x_try = 0, dx = 1e-3):

    x_minus = x_try - dx
    x_plus = x_try + dx

    if f(x_try) - f(x_minus) < 0 and f(x_try) - f(x_plus) < 0:

        dp = round(- np.log(dx) / np.log(10)) # decimal places
        x = round(x_try, dp)
        y = round(f(x_try), dp)
        return tuple([x, y])
    
    elif f(x_try) - f(x_minus) > 0:
        return findmin(f, x_minus, dx)
    
    elif f(x_try) - f(x_plus) > 0:
        return findmin(f, x_plus, dx)
    
    else:
        raise Exception("something went wrong")
```
