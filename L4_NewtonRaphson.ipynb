### Newton Raphson Method

**Theory**

A better estimation of root of $f(x)$ from previous estimation is given by

$$
x_{n+1}=x_n - \frac{f(x_n)}{f'(x_n)}
$$

*Proof* 
$$
f'(x_n) = \frac{f(x_{n+1}) - f(x_{n})}{x_{n+1}-x_n}
$$
Because $x_{n+1}$ is defined as a better estimation to root, $f(x_{n+1})\approx0$. Hence,
$$
x_{n+1}=x_n - \frac{f(x_n)}{f'(x_n)}
$$
**Q.E.D**

To cease the algorithm, there are two ways :
1. if&nbsp; $|f(x_n)| <$ tolerance
2. if&nbsp; $|x_{n+1}-x_n|<$ tolerance

```python
# algorithm 1
'''to test the function whether has root, can set tolerance = 1e-4 
to test whether the algorithm runs to x -> ± inf'''

"If the function has no real root, the newton raphson line will be flatened at stationary point, hence leading to next x-> ± inf"

import numpy as np

def f(x):
    return x * np.exp(x) - 1

def df(x):
    # manually take derivative
    return np.exp(x) + x * np.exp(x)

def newraph(f, x, tolerance = 1e-4):

    while abs(f(x)) > tolerance:

        if abs(x) > 1/tolerance :
            raise Exception("Encountered stationary point, x -> +-inf")
        
        x -= f(x)/df(x)
        
    return x
```

```python
newraph(f, 0.1, tolerance = 1e-6)
```

##### Version 2

```python
# algorithm 2

def newton_raphson(f, df, x, tolerance = 1e-6, n=0, n_max=1000):
# f: func | df: 1st-derivative | x: guess

    x_new = x - f(x) / df(x)
    n += 1

    if abs(x_new - x) < tolerance:
        return x, n
    else:
         return newton_raphson(f, df, x_new, tolerance, n, n_max)
    
import numpy as np
def f(x):
    return x * np.exp(x) - 1

def df(x):
    return np.exp(x) * (x + 1)

newton_raphson(f, df, 0.1)
```
