### Golden Ratio

Golden ratio $\varphi=\frac{b}{a}$ is defined as:\
$a<b$\
$\frac{a+b}{b}=\frac{b}{a}=\varphi$\
Hence,\
$\varphi = 1 + \frac{1}{\varphi}$

```python
phi = 1
while True:
    phi_new = 1 + 1/phi
    if abs(phi_new - phi)<1e-17:
        print(phi_new)
        break
    else:
        phi = phi_new    
    
```
