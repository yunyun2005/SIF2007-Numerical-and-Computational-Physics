```python
def bisection(f, a, b, tol=1e-8):
    fa = f(a)
    fb = f(b)

    if fa * fb > 0:
        raise ValueError("No root in the interval")

    while (b - a) / 2 > tol:
        mid = (a + b) / 2
        fm = f(mid)

        if fm == 0:
            return mid
        elif fa * fm < 0:
            b = mid
            fb = fm
        else:
            a = mid
            fa = fm

    return (a + b) / 2
```
