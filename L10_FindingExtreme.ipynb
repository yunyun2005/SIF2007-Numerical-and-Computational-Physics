### Finding Extreme

##### Newton Raphson algorithm
To find where $f'(x)=0$ occurs, we iterate
$$
x_{n+1}=x_n - \frac{f'(x_n)}{f''(x_n)}
$$

```python
import numpy as np
def extreme(fp, fpp, x0, tol=1e-6, iter=0): # use newton raphson
    x = x0 - fp(x0) / fpp(x0)
    if abs(x - x0) < tol:
        iter += 1
        return x, fpp(x), iter
    else:
        iter += 1
        return extreme(fp, fpp, x, tol, iter)

def f(x):
    return np.log(x) / x
def fp(x):
    return (1 - np.log(x)) / x**2
def fpp(x):
    return (2 * np.log(x) - 3) / x**3
extreme(fp, fpp, 1, 1e-8)

```
