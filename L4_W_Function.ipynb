### W-Function Calculator using Newton Raphson algorithm

```python
import numpy as np
def Wf(a):
    # implement newton raphson calculator
    def f(x):
        return x * np.exp(x) - a

    def df(x):
        # manually take derivative
        return np.exp(x) + x * np.exp(x)

    def newraph(f, x, tolerance):

        while abs(f(x)) > tolerance:

            if abs(x) > 1/tolerance :
                raise Exception("Encountered stationary point, x -> +-inf")

            x -= f(x)/df(x)
        return x

    # Using built-in newraph to solve xe^x - a
    result = newraph(f, x = a, tolerance = 1e-12)
    return round(result, 5)
    # after tested we found out accuracy is 5~6 when precision is 12
        
```
